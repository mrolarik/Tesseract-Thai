````python
from PIL import Image
import pytesseract

img_path = 'data-text-img/BangkokThai_(vectorized).svg.png'
Image.open(img_path)
````

````python
# Python - using pytesseract for Thai language

txtImg = Image.open(img_path)
text = pytesseract.image_to_string(txtImg, 'tha')

print text
````
